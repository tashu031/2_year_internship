```python
s ={"abc" ,37, False ,  "def" , "ghi", 0 , "abc"}
print(s)   # set does not allow duplicate values
print(type(s))
print(len(s))

s.add(490.98 )   # set add method takes only one argument, so this will raise a TypeError

s2 = {1, 2, 3}
s.update(s2)   # set update method takes an iterable as an argument, so this will add elements of s2 to s
print(s)

s3 = {3, 4, 5}

s4 =s3.union(s2)  # set union method returns a new set that is the union of s and s2, but does not modify s
s5 =s3|s2         # umion
print(s4)
print(s3)   # s3 is not modified by the union method
print(s5)

```

```python
set1  = set()
set2 = set()

for i in range(3):
    i = input("Enter a number: ")
    set1.add(i)
print(set1)

for i in range(3):
    i = input("Enter a number: ")
    set2.add(i)


print(set2)
s3 = set1.intersection(set2)
s4 = set1 & set2
print(s3)   
print(s4)   # intersection of set1 and set2

s5 = set1.difference(set2)
s6 = set1 - set2
print(s5 ,s6)   # difference of set1 and set2

print(s3^s4)   # symmetric difference of set1 and set2




```
