```python

import pandas as pd
df1 = pd.DataFrame({
    'ID' : [12,2,5,7,3],
    'name' : ['A1','A5','A6','A2','A4'],
    'sub'  : ['s1' ,'s3','s4','s6','s2']
    
})

df2 = pd.DataFrame({
    'ID' : [10,2,78,3,5],
    'Class'  : ['C1','C3','C4','C2','C5'],
    'marks'  : [12,56,8,9,4]
})

result = df1.merge(df2 , on='ID' ,how = 'inner')
print(f"\nInner merge :\n")
print(result)

```

```python
import pandas as pd
one = {
    'id' :[1,3,5,6,8],
    'name' : ['a1','b1','c2','a4','a6'],
    'subject'  : ['sub1','sub3','sub3' ,'sub2','sub4']
    
 }

df1 = pd.DataFrame(one)

two = {
    'id' : [3,4,6,7],
    'name' : ['b1','b1','c2','a1'],
    'subject'  : ['sub5','sub3','sub2' ,'sub1']
    }

df2 = pd.DataFrame(two)
res = df1.merge(df2 , on=['id', 'name'] , how = 'inner')
print(res)
```

```python
import pandas as pd
one = {
    'ID' : [10,2,3,5],
    'name' : ['a1','b1','c2','a4'],
    'subject'  : ['sub1','sub3','sub3' ,'sub2'],
    'marks' : ['m1' ,'m3','m4','m2']
 }

df1 = pd.DataFrame(one)

two = {
    'ID' : [12,3,2 ,1],
    'name' : ['a2','b1','c2','a1'],
    'subject'  : ['sub3','sub3','sub2' ,'sub1'],
    'marks' : ['m2' ,'m3','m7','m2']
 }

df2 = pd.DataFrame(two)

res = pd.concat([df1,df2] , ignore_index=True)
print(res)


df3 = pd.DataFrame({
    'ID' : [10,2,78,3,5],
    'Class'  : ['C1','C3','C4','C2','C5'],
})

result = res.merge(df3 , on ='ID'  , how = 'outer')
print(result)
```
