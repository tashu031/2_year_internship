```python
# 1) Replace Nan with 0 and Interchange 3 rows and 3 columns of 2D array [[6, -8, 73, -110], [np.nan, -8, 0, 94]]

import numpy as np
arr  = np.array(
    [[6, -8, 73, -110], [np.nan, -8, 0, 94]]
)

print("Original array :")
print(arr)
# replace Nan with 0
arr = np.nan_to_num(arr)
print(f"\nArray after replace nan :\n{arr}")

# interchange 3 rows nd 3 columns 

arr[[0,1],:]  = arr[[1,0],:]
print(f"\nAfter  Row interchange:\n{arr}")

arr[:,[0,1,2]] = arr[:,[2,0,1]]
print(f"\nafter column interchange :\n{arr}")

```

```python
# 2) Move axes of 3D array to new positions
arr = np.array([
    [[3,5,6],[5,6,8],[6,8,9]],
    [[34,56,7] ,[2,5,2],[2,6,90]]
])
print(arr)
# move axes of 3D array to new position :
new_arr = np.moveaxis(arr ,[0,1,2] , [2,0,1])
print(new_arr)
```

```python
#) Replace NaN values with average of columns 

arr  = np.array (
    [[3,5,7] ,[4,np.nan ,6] ,[8,2,np.nan]]
)
print("Original array :")
print(arr)

# replace nan with avg of col
col_mean = np.nanmean(arr , axis =0)
print("\n new array after replace nan with avg of columns :")
new_arr = np.where(np.isnan(arr) ,col_mean ,arr)
print(new_arr)
```

```python
# 4) Replace negative value with zero in numpy array using replace 

arr = np.array([
    [2,4,6] ,[-4,6,3] ,[-3,7,-9]
])
print("Original array :")
print(arr)

# replace neg value with zero 
res = np.where(arr<0 ,0,arr)
print(f"\n new array after replace negetive value \n{res}")
```

```python
#  Calculate average mean median mode values of two NumPy 2d-arrays 
import numpy as np
arr1 = np.array([[3,5,7],[2,4,6]])
arr2 = np.array([[3,68,9],[45,7,9]])



avg = (arr1 + arr2) / 2

print("Average Array:")
print(avg)

com_arr = np.concatenate((arr1.flatten(), arr2.flatten()))

mean = np.mean(com_arr)
median = np.median(com_arr)

values, counts = np.unique(com_arr, return_counts=True)
mode = values[np.argmax(counts)]

print("\nMean =", mean)
print("Median =", median)
print("Mode =", mode)

```

```python
# 6) Solve the following equation using linalg() and inverse Matrix method 
# x - 2y + 3z = 9 
# -x + 3y - z = -6 
# 2x - 5y + 5z = 17

import numpy as np

a = np.array([[1,-2,3],[-1,3,-1],[2,-5,5]])
b = np.array([9,-6,17])

x = np.linalg.solve(a,b)
print(x)
print(f"the solution of the euations : \nx = {x[0]} \ny = {x[1]}\nz = {x[2]}")
```
