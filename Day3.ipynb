```python
# if else 
m = int(input("Enter a number: "))
if(m >=90):
    print(" Grade A+")
elif( m>=80):
    print(" Grade A")
elif(m>=70):
    print(" Grade B")
elif(m>=60):
    print(" Grade C")
else:
    print("Fail")

```

```python
list2 = [3,5,3,6,3,3,6,8,8,3]
l = ["python" , "java" , "c++" , "C#"]
print(list2[::-1])
l.reverse()
print(l)
l.append("javascript")
print(l)


list2.append(9)      #only one element can be added at a time
print(list2)

list2.insert(3,12)      # insert 12 at index 3
print(list2)
list2.remove(3)        # removes first occurrence of 3
print(list2)

list2.pop()
print(list2)
print(len(list2))

l.append(list2)
print(l)
print(l[5][6])

l.extend([34,56,8,5])      # adds multiple elements to the end of the list
print(l)
```

```python
list2 = [100, 2, 3,409, 8 , 34 , 89 , 35 , 4, 5]
sum =0
for x in list2:
    sum +=x
print(f"Sum is {sum}")    

max = list2[0]
for x in list2:
    if x>max :
        max =x
print(f"Max is {max}")
print(f"Min is {min(list2)}")

```
