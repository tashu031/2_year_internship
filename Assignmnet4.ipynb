```python
#1. Write a Python function to find the maximum of three numbers.

n1 = int(input("enter first no . :"))
n2 = int( input("enter second no."))
n3 = int( input("enter third no."))

def max(n1,n2,n3) :
    if(n1>=n2 and n1>=n3):
        print(f"{n1} is Max.")
    elif(n2 >=n1 and n2>=n3):
        print(f"{n2} is max.")
    else:
        print(f"{n3} is max.")    


max(n1,n2,n3)
```

```python
#2. Write a Python function that takes a list and returns a new list with distinct elements from the first list.

l= [2,3,2,3,4 ,3  ,6 ,7 ,7 ,5,4]

def dist_list (l):
    
    new_l =[]
    
    for i in l: 
        
        if( i not in new_l):
            new_l.append(i)
    return new_l        

d_list  = dist_list(l)
print("distinct List :" , d_list)
```

```python

# 3. Write a Python function to multiply all the numbers in a list.

l =[4,2,5]

def mul_list(l):
    mul =1
    for i in l:
        mul = mul*i
    return mul

multi_l = mul_list(l)
print(f"Multiplication of all the elements are : {multi_l}")
```

```python
#4. Write a Python function to calculate the factorial of a number (a non-negative integer). The function accepts the number as an argument.


n = int(input("Enter a number : "))
def factorial(n):
    fac =1
    for i in range(1,n+1):
        fac  = fac*i
    return fac

print(f"Factorial of the {n} is {factorial(n)}")    
```

```python
#5. Write a Python program to reverse a string.

s = input("enter a string :")

def rev_s(s):
    rev_str = s[::-1]
    return rev_str

print(f"reverse of the {s} is {rev_s(s)}")
```

```python
# 7. Write a Python function to Print Even Numbers from a Given List


l = [2,3,4 ,6 ,5,1, 78 ,32]

def print_even(l) :
    for i in l:
        if(i%2 ==0):
            print(i)
        else:
            continue

print_even(l)            
```

```python
# 8. Write a Python function that takes a number as a parameter and checks whether the number is prime or not.

num = int( input("enter a number :"))

def prime(n):
    count=0
    for i in range( 2, (n//2)+1):
        if(n % i ==0):
            count+=1
        else:
            continue
    return count
    
if(prime(num)==0):
    print(f"{num} is Prime.")
else:
    print(f"{num} is not Prime.")

```

```python
# 9. Write a Python function that accepts a string and counts the number of upper and lower case letters.

s = input("Enter your string :")

def count(s):
    countX =0
    countx =0
    for ch in s :
        if(ch>="A" and ch<="Z"):
            countX+=1
        elif(ch>='a' and ch<='z')  :
            countx+=1  
    return countX ,countx

c = count(s) 
print(f"Upper Case in {s} : {c[0]}")  
print(f"Lower Case in {s}: {c[1]}")     
```
