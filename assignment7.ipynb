```python
import pandas as pd

emp = [
    {'id' : 101 ,'name' : 'Tashu' , 'salary' : 50000} ,
    {'id' : 103 ,'name' : 'priya' , 'salary' : 20000} ,
    {'id' : 105 ,'name' : 'kiran' , 'salary' : 90000} ,
    {'id' : 106 ,'name' : 'mahi' , 'salary' : 35000} ,

]

df = pd.DataFrame(emp)

print(df)

print("\nemployee with salary >= 50000  - ")
res = df[df['salary'] >= 50000]
print(res)

print('\nLimited rows selection with given column')
print(df.loc[2:, 'name' ])


```
