```python
import numpy as np

arr = np.array([3,5,7,9,20,10])
arr2d = arr.reshape(2,3)

print(f"1D array : \n{arr}")
print(f"\nAfter conversion in 2-D : \n{arr2d}")

print("\n\nArray Attributes:")
print("Shape:", arr2d.shape)
print("Dimension:", arr2d.ndim)
print("Data Type:", arr2d.dtype)
print("Item Size:", arr2d.itemsize)
```

```python
# Convert a Python list into a NumPy array

l = [10,20,70,80,90]

arr = np.array(l)
print(f"List to numpy Array :\n{arr}\nType : {type(arr)}")
```

```python
# Reverse a 1D NumPy array

arr = np.array(
    [2,4,6,9,0,10,7]
)

rev_arr = arr[::-1]
print("Original array :")
print(arr)
print("\nReverse array :")
print(rev_arr)
```

```python
# Create a 4×4×3 array and extract value at its second set, first row and last column
arr = np.arange(48).reshape(4, 4, 3)
print("\n4x4x3 Array:")
print(arr)


print("\nSecond Set, First Row, Last Column:")
print(arr[1,0,2])


#  Slice first two rows and first two columns of second set from 4×4×3 array
print("\nFirst Two Rows and First Two Columns of Second Set:")
print(arr[1, :2, :2])


```

```python
# 8. Create a 4×4 and Extract Odd Rows and Even Columns
arr = np.arange(1, 17).reshape(4, 4)

print("\n4x4 Array:")
print(arr)
print("\nOdd Rows and Even Columns:")
print(arr[0::2, 1::2])
```

```python
# 10. Replace all odd numbers with -1 using for loop
arr = np.array([[23, 56, 78, 93],
                [71, 82, 13, 24]])

for i in range(arr.shape[0]):
    for j in range(arr.shape[1]):
        if arr[i][j] % 2 != 0:
            arr[i][j] = -1

print("\nReplace Odd Numbers with -1:")
print(arr)

```

```python
# 11. Get indices of non-zero elements
arr = np.array([1, 0, 2, 0, 3, 0, 4])

print("\nIndices of Non-Zero Elements:")
print(np.nonzero(arr))
```

```python
# 12. Arithmetic Operations
arr1 = np.array([10, 20, 30])
arr2 = np.array([1, 2, 3])

print("\nAddition:")
print(arr1 + arr2)

print("\nMultiplication:")
print(arr1 * arr2)

```

```python
# 13. Dot Product
arr1 = np.array([15, 20, 25])
arr2 = np.array([10, 40, 37])

dot_product = np.dot(arr1, arr2)

print("\nDot Product:")
print(dot_product)
```
