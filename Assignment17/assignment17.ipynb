```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("assignment17").getOrCreate()
```

```python
df = spark.read.csv(  "sales.csv", header=True, inferSchema=True)
df.show()
```

```python
sort_df = df.orderBy(df.sales.desc())
```

```python
# sort all products by sales in descending order and display the results,
print("Sorted Data: ");
sort_df.show()
```

```python
# display the top 3 products with the highest sales values,

top_3 = sort_df.limit(3)
print("top 3 high Sales product: ");
top_3.show()
```

```python
#  filter products with sales greater than 80,000 and save the output as a CSV file

df2 = sort_df.filter(df.sales>80000)
print("products with sales greater than 80,000 :");
df2.show()
```

```python
df2.write.mode("overwrite").csv("output/high_sales_products",header= True)
```

```python
spark.stop()
```
