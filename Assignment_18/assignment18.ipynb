```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col
```

```python
spark = SparkSession.builder.appName("Assignment18").getOrCreate()
```

```python
df = spark.range(0,5000000).withColumn("value",col("id")%1000)
```

```python
df.show(5)
```

```python
print("Before Partitions :",df.rdd.getNumPartitions())
```

```python
df.write.mode("overwrite").csv("output/before_partitions_data",header = True)
```

```python
df_rep = df.repartition(12)
```

```python
print("After Partitions :",df_rep.rdd.getNumPartitions())
```

```python
df_rep.write.mode("overwrite").csv("output/after_partitions_data",header = True)
```

```python
df_coalesce = df_rep.coalesce(3)
```

```python
print("After Coalesce Number of partitions :",df_coalesce.rdd.getNumPartitions())
```

```python
df_coalesce.write.mode("overwrite").csv("output/after_coalesce_data",header=True)
```
