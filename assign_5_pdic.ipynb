```python
# 4) Write a Python program to calculate the total, highest value, 
#       and lowest value from a tuple of integers without using the built-in sum(), max(), and min() functions

tup = (2,4,6,8,3)
sum =0
max = tup[0]
min= tup[0]

for element in tup :
    sum =sum + element 
    if element > max :
        max = element
    if element <min:
        min = element

print(f"sum of tuple :{sum}")
print(f"max element :{max}")
print(f"min element :{min}")

```

```python
# 5) Write a Python program to filter a tuple . n = (3, 14, 7, 22, 9, 41, 18, 5), keep only values greater than 10 

n = (3, 14, 7, 22, 9, 41, 18, 5)
temp =[]
for e in n :
    if(e>10):
        temp.append(e)
filter_tup = tuple(temp)
print(filter_tup)

```

```python
#  7) Write a Python program to combine two sets into one, containing all unique elements from both sets.

s1 = {23,5,6,8,2,3}
s2 = {4,6,8,11,78}

s3 = s1 ^ s2
print(f"New set with all unique elements from both set : {s3}")
```
