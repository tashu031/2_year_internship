```python
import numpy as np

arr1 = np.array([
    [10, 20],
    [30, 40]
])

arr2 = np.array([
    [50, 60],
    [70, 80]
])

avg = (arr1 + arr2) / 2

print("Average Array:")
print(avg)

com_arr = np.concatenate((arr1.flatten(), arr2.flatten()))

mean = np.mean(com_arr)
median = np.median(com_arr)

values, counts = np.unique(com_arr, return_counts=True)
mode = values[np.argmax(counts)]

print("\nMean =", mean)
print("Median =", median)
print("Mode =", mode)
```
