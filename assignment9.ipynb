```python
import pandas as pd
df  = pd.read_csv('regex_dataset.csv')
# print(df)

# print(df.isnull().sum())
# print(df.info())
# print(df.shape)
# print(df.columns)

df['salary'] = df['salary'].str.replace(r'\D+',"" ,regex=True)
print(df)
df['name'] = df['name'].str.replace(r'[^A-Za-z]','',regex = True)
print(df)

df['valid_email'] = df['email'].str.contains(r'^[A-Za-z0-9._%+-]+@(gmail\.com|yahoo\.in|outlook\.com)$')
print(df[['email','valid_email']])


print(df)
```

```python
import pandas as pd

df = pd.read_csv("datetime_dataset.csv")
print(df)

df['joining_date'] = pd.to_datetime(
    df['joining_date'],
    errors='coerce'
)

# now datetime functions will work
df['year'] = df['joining_date'].dt.year

df['month'] = df['joining_date'].dt.month

print(df)
print(df.dtypes)

current_time = pd.Timestamp.now()
print(current_time)


print("\nEmployees Joined In 2025")

print(
    df[df['joining_date'].dt.year == 2025]
)


print("\nSorted Dataset")
print(df.sort_values(by='joining_date'))


```

```python
import pandas as pd

df = pd.read_csv("tips.csv")
print(' Display First 5 rows.....')
print(df.head())
print(df.info())

print(f"check the missing values : \n{df.isnull().sum()}")
print(f"Count duplicate rows:\n{df.duplicated().sum()}")


print(df['day'].head())
print("only only total_bill and tip column ")
print(df[['total_bill','tip']].tail())


# print the data with total_bill is >=20
res = df[df['total_bill']>=20]


# print only Male's data
res = df[df['sex']=='Male']


#  sort the data with the higher total_bill and then tip
res = df.sort_values(by =['total_bill' , 'tip'] , ascending=[True,False])


print(res)




df['sex'] = df['sex'].replace({
    'Male' : 0,
    "Female" : 1
})

df['smoker'] = df['smoker'].replace({'Yes':1 ,"No": 0})
print(df)


#  data grouping 
res2 = df.groupby('day')['tip'].sum()          # total tip day wise
res2 = df.groupby('day')['tip'].mean()
print(res2)

res3 = df.groupby(['day' , 'time'])['sex'].count()           # count no of customers per day and dinner and lunch wise
print(res3)
res4 = df.groupby(['day' , 'time','sex'])['sex'].count()
print(res4)


```
