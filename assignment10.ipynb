```python
import numpy as np

arr = np.array(
    [[6, -8, 73, -110],  [np.nan, -8, 0, 94]]
    )

print("Display Original Array:")
print(arr)

# Replace NaN with 0
arr = np.nan_to_num(arr,nan=0)

print("\nAfter Replacing nan with 0:")
print(arr)

# Interchange rows and columns..............
a= arr.T

print("\nAfter Interchanging Rows and Columns:")
print(a)
```
